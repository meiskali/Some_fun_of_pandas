### Step 4. Discover what is the mean age per occupation

```python
users.groupby(["occupation"])["age"].agg(["mean"])
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users.groupby(["occupation"])["gender"].agg(["value_counts"])
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(["occupation"])["age"].agg(["min","max"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(["occupation","gender"])["age"].agg(["mean"])
```

### Step 8.  For each occupation present the percentage of women and men

```python
tt=users.groupby(["occupation"])["gender"].agg(["value_counts"]).reset_index()
tt["Total"]=0
tt
```

```python
for i,y in enumerate(tt["occupation"]):
    total=tt[tt.isin([y]).any(1)]["value_counts"].sum()
    tt.loc[i,"Total"]=total
```

```python
tt["percentage"]=0
```

```python
tt.head()
```

```python
for i in range(len(tt["Total"])):
    tt.loc[i,"percentage"]=tt.loc[i,"value_counts"]/tt.loc[i,"Total"]*100
```
